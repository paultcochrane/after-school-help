```python
p = plot(sin(x), 0, 2*pi, figsize=(10,4), ticks=pi/4, tick_formatter=pi, axes_labels=['$x$', '$y$'])
p += plot(cos(x), 0, 2*pi, linestyle="--")
p += text(r'$(a)$', (2*pi+0.2, 1))
p += text(r'$(b)$', (2*pi+0.2, 0.1))
p.save('basic-sine-curves.pdf')
p.show()
```

```python
a = var('a')
b = var('b')
c = var('c')
d = var('d')

f = a*sin(b*(x - c)) + d

args = dict(a=2, b=1, c=pi/2, d=-1)
p = plot(f(**args), -pi, 2*pi, ticks=pi/4, tick_formatter=pi, axes_labels=['$x$', '$y$'])
p += text(r'$(c)$', (2*pi+0.2, f(**args, x=2*pi)+0.1))
print(f(**args))

args = dict(a=1.5, b=3, c=0, d=0)
b_colour = (0, 0.7, 0)
p += plot(f(**args), 0, 2*pi, rgbcolor=b_colour)
p += text(r'$(d)$', (2*pi+0.2, f(**args, x=2*pi)+0.1), rgbcolor=b_colour)
print(f(**args))


args = dict(a=1, b=0.5, c=0, d=2)
c_colour = (0.7, 0, 0)
p += plot(f(**args), -pi, 3*pi, rgbcolor=c_colour, figsize=(10,4))
p += text(r'$(e)$', (2*pi+0.2, f(**args, x=2*pi)+0.1), rgbcolor=c_colour)
print(f(**args))

p.save('sine-curves-1.pdf')
p.show()
```

```python
a = var('a')
b = var('b')
c = var('c')
d = var('d')

f = a*sin(b*(x - c)) + d

args = dict(a=1, b=1, c=pi/2, d=0)
p = plot(f(**args), -pi, 2*pi, ticks=pi/4, tick_formatter=pi, axes_labels=['$x$', '$y$'])
p += text(r'$(f)$', (2*pi+0.2, f(**args, x=2*pi)+0.1))
print(f(**args))

args = dict(a=0.5, b=2, c=-pi/2, d=-0.5)
b_colour = (0, 0.7, 0)
p += plot(f(**args), -pi/2, 2*pi, rgbcolor=b_colour)
p += text(r'$(g)$', (2*pi+0.2, f(**args, x=2*pi)+0.1), rgbcolor=b_colour)
print(f(**args))

args = dict(a=1.5, b=4, c=pi/4, d=1)
c_colour = (0.7, 0, 0)
p += plot(f(**args), 0, 2*pi, rgbcolor=c_colour, figsize=(10,4))
p += text(r'$(h)$', (2*pi+0.2, f(**args, x=2*pi)+0.1), rgbcolor=c_colour)
print(f(**args))

p.save('sine-curves-2.pdf')
p.show()
```
