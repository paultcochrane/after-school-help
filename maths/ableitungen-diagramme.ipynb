```python
import matplotlib.pyplot as plt

tangent_colour = (0, 0.7, 0)
point_colour = (0.8, 0, 0)

x = var('x')

f = x^2 + 2*x - 3
p = plot(f, x, -6, 4, axes_labels=['$x$','$y$'])
p += plot(point((2, 5), rgbcolor=point_colour))
p += line([(1, -1), (2, 5), (3, 11)], rgbcolor=tangent_colour)
p += text(r'$\mathrm{Punkt}$', (3, 5), rgbcolor=point_colour)
p += arrow2d(path=[[(3, 4), (2.5, 2), (2, 4.8)]], rgbcolor=point_colour)
p += text(r'$\mathrm{Tangente}$', (4, 10), rgbcolor=tangent_colour)
p += arrow2d(path=[[(4, 9), (3.5, 7), (2.75, 9)]], rgbcolor=tangent_colour)

p += plot(point((-2, -3), rgbcolor=point_colour))
p += line([(-3, -1), (-2, -3), (-1, -5)], rgbcolor=tangent_colour)
p += text(r'$\mathrm{Punkt}$', (-3, -3), rgbcolor=point_colour)
p += arrow2d(path=[[(-3, -4), (-2.5, -5), (-2, -3.2)]], rgbcolor=point_colour)
p += text(r'$\mathrm{Tangente}$', (-3, -5), rgbcolor=tangent_colour)
p += arrow2d(path=[[(-3, -6), (-2, -8), (-1.25, -5)]], rgbcolor=tangent_colour)

p.save('quadratic-with-tangents.pdf')
p.show()
```
