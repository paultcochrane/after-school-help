# A fisherman's fastest path from shore to boat

Consider figure below. The distance from the fisherman standing on the edge of the shore to the boat is 10 m, the distance from the edge of the shore to the boat is 6 m. Using Fermat's principle, determine the shortest time it will take the fisherman to reach the boat if the speed of the fisherman's movement along the shore is 3 m/s, and the lake is three times less.

![](fisherman-boat-fastest-path.jpeg)

## Solution

There are a few ways that the fisherman could reach the boat.  He could go directly to the boat through the water, travelling 10 m to the boat.  Or he could travel along the shore until the distance in the water is shortest (i.e. 6 m) and then he could go through the water the rest of the way.  Neither of these options is optimal: in one case he doesn't make any use of the speed advantage along the shore and in the other case he travels too far along the shore and ends up choosing a much longer distance than necessary.  The best path lies between these two extremes and that's what we have to find.

Let's collect some information from the question that we are told and that we can deduce in order to help us solve the problem.  We're told that his speed along the shore is 3 m/s.  Let's denote this speed as $v_s$.  The speed in the water we can denote by $v_w$.  We're told that the speed in the water is three times less than the speed along the shore, hence $v_w$ is 1 m/s.  From the geometry of the situation we can see that the paths in the water and the path along the shore make a scaled 3-4-5 right-angled triangle.  Doubling the side lengths of a 3-4-5 right-angled triangle gives 6-8-10 (you can verify that this is still a right-angled triangle because $6^2 + 8^2 = 10^2$). In other words, the path along the shore is 8 m.

Now that we have deduced most of the information we can from the diagram and the question, the task is now to find the best path from the fisherman's position on the shore to the boat.  Note that this is *not* the shortest path.  It is, however, the path that takes the least *time*, which is where Fermat's principle comes in.  If we change our perspective slightly, we can consider the problem to involve movement at a given velocity in one medium (along the shore) and then movement at another velocity in a difference medium (in the water).  Thus, what we're really looking at is Snell's law in ray optics.  This becomes clearer if we consider the path a light ray would take from the boat to the fisherman's initial position.  Why look at the problem from this perspective?  It turns out that the physics of finding the path taking the least time is independent of direction: i.e. the path *from* the boat is equivalent to the path *to* the boat.  Considering the perspective *from* the boat we can see that a ray travelling from the boat and then *along* the shore will give us what is known as the *critical angle*.  This is the minimum angle that would keep a light ray within the given medium.  I.e. a calculation using Snell's law at the critical angle is equivalent to the problem we're trying to solve with the fisherman getting to the boat in the least time.

Let's turn the situation on its head to see what's happening in terms of Snell's law from the perspective of a light ray travelling from the boat. Consider the following diagram:

![](light-ray-from-boat.png)

Here we have a ray travelling from the red point in the blue region at an angle $\theta_w$ to the normal to the white region.  It is then bent at this boundary and travels along the interface to the second red point.  The bent ray travels at an angle $\theta_s$ to the normal (i.e. a line perpendicular to the boundary between the two regions).  Using Snell's law, we have the following relationship

$$
\frac{\sin \theta_s}{\sin \theta_w} = \frac{v_s}{v_w}
$$

We know that $v_s$ is 3 m/s, $v_w$ is 1 m/s and that $\theta_s$ is $90^\circ$ (equivalently: $\pi/2$ radians).  Since $\sin(\pi/2) = 1$, we can simplify the expression above to:

$$
\frac{1}{\sin \theta_w} = \frac{3}{1}
$$

Rearranging in terms of $\sin \theta_w$, we have

$$
\sin \theta_w = \frac{1}{3}
$$

Taking the inverse sine, we have an expression for $\theta_w$:

$$
\theta_w = \sin^{-1}\left( \frac{1}{3} \right)
$$

This is the angle to the normal at the boundary that a light ray would take along the path of least time between the two red points.  We can now work out the distance travelled both in the water and along the shore.  Once we have that information, we can use the velocities in each medium to work out the time taken along this path and hence calculate the time that the fisherman would take along the best path from his position on the shore to the boat.

From geometry, we see that the angle between the vertical line and the ray in the blue region at the red dot is equal to $\theta_w$.  Thus, we can calculate the length $d_w$, which is the hypotenuse of the right-angled triangle with sides 6 m and $d_{s'}$:

$$
d_w = \frac{6}{\cos \theta_w}
$$

From the diagram, we know that the length along the shore (along the boundary between the blue and white region) is 8 m.  Thus, the distance the ray travels along the shore, $d_s$, plus the remaining distance along the shore $d_{s'}$ is equal to 8 m, i.e. we have this relation:

$$
d_s + d_{s'} = 8
$$

which means we have an expression for $d_s$ in terms of $d_{s'}$

$$
d_s = 8 - d_{s'}
$$

Again, from geometry, we can work out what $d_{s'}$ is in terms of $\theta_w$ by using the tangent and the adjacent side ($d_{s'}$ is the opposite side of the triangle):

$$
d_{s'} = 6 \tan \theta_w
$$

Now we know how far the distance along the shore is in terms of the angle $\theta_w$:

$$
d_s = 8 - 6 \tan \theta_w
$$

We're now left with the task of finding out how long it takes to travel each of the two components of the path, $d_w$ and $d_s$.  Since the speed in each medium is constant, the distances can be written in terms of their velocities and times, i.e.:

$$
d_w = v_w t_w\\
d_s = v_s t_s
$$

where $t_w$ and $t_s$ are the times in the water and along the shore, respectively.  These can be written in terms of distance and velocity like so:

$$
t_w = \frac{d_w}{v_w}\\
t_s = \frac{d_s}{v_s}
$$

The total time for the best path is

$$
t = t_s + t_w = \frac{8 - 6 \tan \theta_w}{v_s} + \frac{6}{v_w \cos \theta_w}\\
= \frac{8 - 6 \tan \theta_w}{3} + \frac{6}{\cos \theta_w}
$$

where I've substituted the values of the speed along the shore and the speed in the water into the equation.

We know that

$$
\theta_w = \sin^{-1} \left( \frac{1}{3} \right) \approx 0.3398
$$

Subsituting this value into the equation for $t$ above, we get

$$
t = t_s + t_w \approx 1.9595 + 6.3640 \approx 8.3235 \mathrm{s}
$$

In other words, the shortest time for the fisherman to travel from his initial position to the boat is 8.3235 s.

## Numerical calculations

Below is the calculation I used to calculate the numerical values of $t_s$, $t_w$ and hence $t$ in the section above.

```python
theta_w = asin(1/3.)
ts = (8 - 6*tan(theta_w))/3
tw = 6/cos(theta_w)
t = ts + tw
print(f"ts = {ts.n():0.5}; tw = {tw.n():0.5}; t = {t.n():0.5}")
```

## Extension

Now, because I'm me, and I can't simply solve something in one way and be done with it, I decided to plot the total time for paths corresponding to different $\theta_w$ angles, ranging from 0 (the fisherman goes 8 m along the shore and then 6 m in the water) to $theta_{\mathrm{max}} = \cos^{-1}(3/5)$ (the fisherman follows the most direct path, 10 m only in the water).  I wanted to see this plot because it gives a more intuitive feel for how the time is a minimum between the two extremes.

I define a special (symbolic) variable `theta_w`, I then define the time along the shore `ts` and the time in the water `tw` in terms of `theta_w`.  This allows me to write the total time `t`, which is then a function of `theta_w`.  I calculate the maximum value `theta_w` can take, corresponding to the direct path from the fisherman only through the water to the boat, `theta_max`.  Then I plot `t` as a function of `theta_w`.

```python
theta_w = var('theta_w')
ts = (8 - 6*tan(theta_w))/3
tw = 6/cos(theta_w)
t = ts + tw
theta_max = acos(3/5)

plot(t, (theta_w, 0, theta_max), axes_labels=['$\\theta_w$','$t$'])
```

The graph shows the time taken for `theta_w` equal to zero.  This is the same as travelling 8 m at 3 m/s and then 6 m at 1 m/s.  In other words

$$
\frac{8}{3} + 6 \text{s} \approx 8.667 \text{s}
$$

At the largest angle, the time taken is 10 s because this is the time taken to travel a distance 10 m travelling at 1 m/s.

Reading the graph, we can approximate the lowest time to be at a `theta_w` value of roughly 0.35 and is approximately 8.3 s.  These values correspond to those we got for $theta_w$ and $t$ in the solution section above.
