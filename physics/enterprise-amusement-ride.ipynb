# The Enterprise amusement park ride

## Question

[The Enterprise](https://en.wikipedia.org/wiki/Enterprise_(ride)) is an amusement ride sometimes found at carnivals and amusement parks. It consists of a large wheel-like structure with gondolas attached at the wheel's rim, similar to a Ferris wheel. The wheel initially starts out horizontal (the plane of the wheel being parallel to the plane of the ground) however, after a certain rotational speed has been reached, the wheel lifts to the vertical (the plane of the wheel being perpendicular to the plane of the ground). This means that the gondolas at the top of the wheel are upside down.

Consider such an amusement ride with a diameter of 22m. Calculate both algebraically and numerically the minimum speed, $v_\mathrm{min}$, that this ride must have so that a passenger in a gondola at the wheel's highest point does not fall onto the gondola's roof.

## Answer

The force keeping the passengers on a circular path is the [centripetal force](https://en.wikipedia.org/wiki/Centripetal_force).  A passenger in a gondola experiences the feeling of being pushed outwards, away from the wheel's hub, and into the floor of the gondola.  The force that the passengers experience is called the [centrifugal force](https://en.wikipedia.org/wiki/Centrifugal_force), which is a fictitious force experienced by objects within rotating systems and acts radially away from the axis of rotation (i.e. outwards, away from the wheel's hub). Since we wish to consider the experience of a passenger in one of the gondolas (i.e. in one of the rotating frames of reference), it is useful (and relevant) in this case to consider the magnitude of the centrifugal force, which we will denote by $F_c$.  The passenger also experiences a force due to gravity (i.e. their weight) which is directed vertically downwards towards the earth; this force we denote by $F_g$.  When the magnitude of the centrifugal force is equal to the weight of the passenger, then they will remain in their gondola at the wheel's highest point and thus won't fall onto the gondola's roof.  In other words, we need to find the velocity when $F_c = F_g$.

We know that the centripetal force is

$$
F_c = m a_c
$$

where $m$ is the mass and $a_c$ is the centripetal acceleration.

The centripetal acceleration is

$$
a_c = \frac{v^2}{r}
$$

where $v$ is the *speed* of the wheel's rim and $r$ is the wheel's radius.

We know that the magnitude of the centripetal force is equal to the centrifugal force experienced by the passenger within the rotating frame of reference, hence we can be lazy and use the same symbol for the centrifugal force $F_c$ and substitute the expression for centripetal acceleration from above into the equation of the force to get

$$
F_c = m \frac{v^2}{r}
$$

for the centrifugal force experienced by the passenger.

The passenger has a weight due to gravity of

$$
F_g = m g
$$

where $g$ is the acceleration due to gravity.

For the passenger to remain on the floor of the gondola at the ferris wheel's highest point, the centrifugal force must equal the passenger's weight, i.e.

$$
F_c = F_g
$$

Substituting the equation for the passenger's weight as well as the equation for the centrifugal force in terms of the wheel's velocity into the above equation we get

$$
m \frac{v_\mathrm{min}^2}{r} = m g
$$

where we've called the wheel's velocity $v_\mathrm{min}$ because this equation represents the situation where the centrifugal force *just balances* the weight force; it's not too big or too little and hence it's the minimum velocity required for the passenger not to fall onto the gondola's roof at the ferris wheel's highest point. 

We can cancel the mass as it appears on both sides of the equation, giving

$$
\frac{v_\mathrm{min}^2}{r} = g
$$

solving this for $v_\mathrm{min}$ we get

$$
v_\mathrm{min} = \sqrt{g r}
$$

Given that we know that $g = 9.81 \mathrm{m/s^2}$ and that $r = 11\ \mathrm{m}$ (the radius is half of the diameter which is 22 m), we can calculate that

$$
v_\mathrm{min} = \sqrt{9.81 \times 11} = 10.388 \mathrm{m/s}
$$

## Derivation of $a_c$

Where did the equation for $a_c$ come from?  There are two ways, and [Physics Ninja explains particularly well how to calculate this](https://www.youtube.com/watch?v=2gNZHt69Fzg).  I'm going to reproduce his explanations here.

### Derivation via a geometric argument

A point on the wheel's rim starts at an initial position $\mathbf{r_i}$ and with an initial *velocity* $\mathbf{v_i}$ (remember, velocity is a vector, hence it has a direction, whereas speed is just the magnitude of the velocity).  After a time $\Delta t$, this point on the wheel's rim has moved to a final position $\mathbf{r_f}$, and now has a final velocity of $\mathbf{v_f}$ (but remember that its speed is still $v$).  The radius vector rotates in the time $\Delta t$ through an angle $\Delta \theta$.  Note that the radius vector $\mathbf{r_i}$ is perpendicular to its corresponding velocity vector $\mathbf{v_i}$ and that the radius vector $\mathbf{r_f}$ is perpendicular to its corresponding velocity vector $\mathbf{v_f}$.  If we draw the radius vectors, we see that there's a vector from the tip of $\mathbf{r_i}$ to $\mathbf{r_f}$ which we'll call $\mathbf{\Delta r}$.  These vectors form a triangle.

TODO: insert picture of radius vector triangle

Since the radius vector $\mathbf{r_i}$ rotates to $\mathbf{r_f}$ making a radius change of $\mathbf{\Delta r}$ in a time $\Delta t$, this gives us a velocity

$$
\mathbf{v} = \frac{\mathbf{\Delta r}}{\Delta t}
$$

The magnitude of this velocity is the speed of a point on the wheel's rim:

$$
v = |\mathbf{v}| = \frac{|\mathbf{\Delta r}|}{\Delta t}
$$

where we take the magnitude only of $\mathbf{\Delta r}$ because $\Delta t$ is a scalar.

We can also draw a similar triangle for the velocity vectors.  We join $\mathbf{v_i}$ to $\mathbf{v_f}$ at their bases and then note that there is a vector from the tip of $\mathbf{v_i}$ to the tip of $\mathbf{v_f}$, which is the change in the two vectors and which we denote as $\mathbf{\Delta v}$, forming a triangle.

TODO: insert triangle of velocity vectors.

Because $\mathbf{v_i}$ is perpendicular to $\mathbf{r_i}$ and $\mathbf{v_f}$ is perpendicular to $\mathbf{r_f}$ we can see that the angle between $\mathbf{v_i}$ and $\mathbf{v_f}$ is $\Delta theta$, hence the triangle for the velocity vectors is similar to the triangle for the radius vectors.  Because these two triangles are similar and isoceles, the ratio of the shortest side to one of the longest sides for the radius vector triangle will be equal to the ratio of the shortest side to one of the longest sides for the velocity vector triangle.  In other words

$$
\frac{|\mathbf{\Delta r}|}{|\mathbf{r_i}|} = \frac{|\mathbf{\Delta v}|}{|\mathbf{v_i}|}
$$

which can be slightly more simply written as

$$
\frac{|\mathbf{\Delta r}|}{r} = \frac{|\mathbf{\Delta v}|}{v}
$$

because $|\mathbf{r_i}| = r$ and $|\mathbf{v_i}| = v$.

We can rearrange this equation for $|\mathbf{\Delta v}|$ like so:

$$
|\mathbf{\Delta v}| = \frac{|\mathbf{\Delta r}| v}{r}
$$

We know that acceleration is

$$
a = \frac{|\mathbf{\Delta v}|}{\Delta t}
$$

hence we can write

$$
a = \frac{|\mathbf{\Delta r}|}{\Delta t} \frac{v}{r}
$$

We know that the speed $v$ is the change in the radius vector, i.e.

$$
v = \frac{|\mathbf{\Delta r}|}{\Delta t}
$$

Substituting this into our equation for $a$, we get

$$
a = v \frac{v}{r} = \frac{v^2}{r}
$$

hence the equation mentioned earlier:

$$
a_c = \frac{v^2}{r}
$$

### Derivation using calculus

The radius vector $\mathbf{r}$ can be split up into its $x$ and $y$ components:

$$
\mathbf{r} = r_x \mathbf{i} + r_y \mathbf{j}
$$

where $r_x$ is the length of the radius vector component in the $x$ direction, $\mathbf{i}$ is the unit vector pointing in the $x$ direction, $r_y$ is the magnitude of the radius vector component pointing in the $y$ direction and $\mathbf{j}$ is the unit vector pointing in the $y$ direction.

Let's denote the angle of the radius vector to the $x$ axis by the variable $\theta$.

We can therefore rewrite the above equation as:

$$
\mathbf{r} = r \cos \theta \mathbf{i} + r \sin \theta \mathbf{j}
$$

where $r$ is the radius of the wheel and thus the magnitude of the radius vector.

We know that the wheel is rotating at a constant speed, hence the radius vector will rotate through the angle $\theta$ within a given time $t$.  Thus we can write the angular velocity of the wheel like so (which, because the wheel's speed is constant, it's angular velocity is also constant):

$$
\omega = \frac{\theta}{t}
$$

Rearranging this equation for $\theta$ we have

$$
\theta = \omega t
$$

Substituting this into our equation for $\mathbf{r}$, we have

$$
\mathbf{r} = r \cos (\omega t) \mathbf{i} + r \sin (\omega t) \mathbf{j}
$$

We know that velocity is the time derivative of distance, hence we can find the velocity vector by differentiating the equation for the radius vector

$$
\mathbf{v} = \frac{d \mathbf{r}}{d t} = -r \omega \sin (\omega t) \mathbf{i} + r \omega \cos (\omega t) \mathbf{j}
$$

Where we've used the property that

$$
\frac{d}{d \theta} \sin \theta = \cos \theta
$$

and

$$
\frac{d}{d \theta} \cos \theta = -\sin \theta
$$

as well as the [chain rule](https://en.wikipedia.org/wiki/Chain_rule).

We can arrive at the acceleration similarly, by differentiating the velocity with respect to time

$$
\mathbf{a} = \frac{d \mathbf{v}}{d t} = -r \omega^2 \cos (\omega t) \mathbf{i} + r \omega^2 \sin (\omega t) \mathbf{j}
$$

which is our centripetal acceleration.  However, we're only interested in the magnitude of this vector $|\mathbf{a}|$ which is the square root of the sum of the squares of its components, i.e.

$$
a_c = |\mathbf{a}| = \sqrt{r^2 \omega^4 \cos^2 (\omega t) + r^2 \omega^4 \sin^2 (\omega t)}
$$

We can extract the common factor of $r^2 \omega^4$ out of the square root to give

$$
a_c = r \omega^2 \sqrt{\cos^2 (\omega t) + \sin^2 (\omega t)}
$$

However, we know from the [trigonometric identity](https://en.wikipedia.org/wiki/List_of_trigonometric_identities#Pythagorean_identities) that

$$
\cos^2 \theta + \sin^2 \theta = 1
$$

hence we can simplify our equation for $a_c$ further to just

$$
a_c = r \omega^2
$$

We know that the angular velocity can be written in terms of the tangential velocity like so:

$$
\omega = \frac{v}{r}
$$

Substituting this value into the above expression for $a_c$ we get

$$
a_c = r \frac{v^2}{r^2} = \frac{v^2}{r}
$$

which is the result we got by using the geometric argument above.
